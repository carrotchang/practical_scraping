## แนะนำเครื่องมือใหม่: Selenium
Selenium จะเปิด web browser ขึ้นมา ให้เราสามารถ interact ได้ เช่น login ใส่ captcha ฯลฯ

```python
from selenium import webdriver
import selenium.webdriver.support.ui as ui

browser = webdriver.Firefox()
browser.get('http://datawarehouse.dbd.go.th/bdw/home/login.html')
wait = ui.WebDriverWait(browser, 30)
wait.until(lambda browser: browser.find_elements_by_class_name('container'))
```

```python
browser.get('http://datawarehouse.dbd.go.th/bdw/est/details1.html?jpNo=0102555000185&jpTypeCode=2')
page = browser.page_source.encode('utf-8')
```
